Light Curve Plot for Exoplanet **K2-2113**(Kepler Telescope)

For More Data on Exoplanets:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('211311380-c05_kepler_Data.xlsx')
df
```

```python
plt.plot(df['A'], df['B'],'b',label='Transit')
#plt.axis([2350,2360,0.994,1.004]) #Zoom-in
plt.xlabel('Time')
plt.ylabel('Stellar luminosity')
plt.legend()
```

```python
plt.plot(df['A'], df['B'],'b',label='Transit')
plt.axis([2350,2360,0.994,1.004]) #Zoom-in
plt.xlabel('Time')
plt.ylabel('Stellar luminosity')
plt.legend()
```
